# Q1 - Chatbot

```python
def chatbot_response(user_input):
    if "admission" in user_input:
        return "Admission process starts in June. Please visit the college website for details."
    
    elif "fees" in user_input:
        return "The fee structure varies by course. Please contact the accounts department."
    
    elif "exam" in user_input:
        return "Exams are conducted at the end of each semester."
    
    elif "timetable" in user_input:
        return "Timetable is available on the student portal."
    
    else:
        return "Sorry, I do not understand your question."


print("Welcome to Student FAQ Chatbot")
print("Type 'exit' to end the chat.")

while True:
    user_input = input("You: ")
    user_input = user_input.lower()

    if user_input == "exit":
        print("Chatbot: Thank you. Goodbye!")
        break

    print("Chatbot:", chatbot_response(user_input))
```

# Q2 - Self Driving Car System

```python
def check_traffic_rule(light_status):
    if light_status == "red":
        return "Stop"
    elif light_status == "yellow":
        return "Slow Down"
    elif light_status == "green":
        return "Go"
    else:
        return "Invalid Traffic Light"


def check_obstacle(obstacle_type):
    if obstacle_type == "pedestrian":
        return "Stop"
    elif obstacle_type == "car":
        return "Avoid"
    elif obstacle_type == "cyclist":
        return "Slow Down"
    elif obstacle_type == "none":
        return "Clear"
    else:
        return "Unknown Obstacle"


def decision_tree(light_status, obstacle_type):
    traffic_action = check_traffic_rule(light_status)
    obstacle_action = check_obstacle(obstacle_type)

    if traffic_action == "Stop":
        return "Final Decision: Stop Vehicle"
    elif obstacle_action == "Stop":
        return "Final Decision: Stop Vehicle"
    elif obstacle_action == "Avoid":
        return "Final Decision: Avoid Obstacle"
    elif traffic_action == "Slow Down" or obstacle_action == "Slow Down":
        return "Final Decision: Slow Down"
    else:
        return "Final Decision: Move Forward"


print("Self-Driving Car Knowledge Base System")

speed_limit = input("Enter Speed Limit: ")
traffic_light = input("Enter Traffic Light (Red/Yellow/Green): ").lower()
crosswalk = input("Is there a pedestrian crosswalk? (Yes/No): ")
obstacle = input("Enter Obstacle (Car/Pedestrian/Cyclist/None): ").lower()

print(decision_tree(traffic_light, obstacle))
```

# Q3 - Medical Diagnosis

```python
def diagnose(s1, s2):
    if "fever" in s1 and "cough" in s2:
        return "\nPossible Diagnosis: FLU"
    elif "fever" in s1 and "rash" in s2:
        return "\nPossible Diagnosis: MEASLES"
    elif "headache" in s1 and "nausea" in s2:
        return "\nPossible Diagnosis: MIGRAINE"
    elif "chest pain" in s1 and "breathlessness" in s2:
        return "\nPossible Diagnosis: HEART DISEASE"
    else:
        return "\nDiagnosis Unclear. Please Consult a Doctor."

print("Medical Diagnosis Expert System")

s1 = input("Enter first Symptom: ").lower()
s2 = input("Enter second Symptom: ").lower()

print(diagnose(s1,s2))
```

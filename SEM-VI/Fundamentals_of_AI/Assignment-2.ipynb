```python
facts = {
    "evening": True,
    "presence": True,
    "temperature_below_18": True,
    "after_9_pm": True
}

def smart_home_system(facts):
    actions = []

    if facts["evening"] and facts["presence"]:
        actions.append("Lights turned ON")

    if facts["temperature_below_18"]:
        actions.append("Heating turned ON")

    if facts["after_9_pm"]:
        actions.append("Thermostat lowered")

    return actions

result = smart_home_system(facts)

print("Smart Home System Actions:")
for action in result:
    print("--", action)
```

```python
facts = {
    "fever": True,
    "cough": True,
    "headache": True,
    "chest_pain": False
}

def medical_diagnosis(facts):
    diagnoses = []

    if facts["fever"] and facts["cough"] and facts["headache"]:
        diagnoses.append("Diagnosis: Flu")

    if facts["fever"] and facts["cough"] and facts["chest_pain"]:
        diagnoses.append("Diagnosis: Pneumonia")

    if facts["cough"] and not facts["fever"]:
        diagnoses.append("Diagnosis: Common Cold")

    return diagnoses


result = medical_diagnosis(facts)

print("Medical Diagnosis Result:")
for d in result:
    print("-", d)
```

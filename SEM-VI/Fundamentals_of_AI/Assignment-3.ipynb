# SET A - Tic Tac Toe

```python
board = [" "]*9
player = "X"

wins = [[0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]]

while True:
    print()
    for i in range(0,9,3):
        print(board[i],"|",board[i+1],"|",board[i+2])
        if i < 6:
            print("--+---+--")
    try:
        move = int(input(f"\nPlayer {player}, enter Position(1-9) : "))-1
        if board[move] != " ":
            print("Position Occupied. Try Again...")
            continue
    except:
        print("Invalid Input. Try Again...")
        continue

    board[move] = player
    if any(all(board[i] == player for i in w) for w in wins):
        print(f"\nPlayer {player} wins!")
        break

    if " " not in board:
        print("\n The Game is a DRAW!")
        break

    player = "O" if player == "X" else "X"

exit()
```

# SET B - 3 Puzzle Game

```python
goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, "_"]
]


state = [
    [1, 2, 3],
    [4, 5, 6],
    ["_", 7, 8]
]


def display_board(board):
    print("\nCurrent State:")
    for row in board:
        print(row)


def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == "_":
                return i, j


def move_tile(board, direction):
    x, y = find_blank(board)


    if direction == "down" and x < 2:
        board[x][y], board[x+1][y] = board[x+1][y], board[x][y]
    elif direction == "up" and x > 0:
        board[x][y], board[x-1][y] = board[x-1][y], board[x][y]
    elif direction == "right" and y < 2:
        board[x][y], board[x][y+1] = board[x][y+1], board[x][y]
    elif direction == "left" and y > 0:
        board[x][y], board[x][y-1] = board[x][y-1], board[x][y]
    else:
        print("Invalid move!")


def is_goal(board):
    return board == goal_state


# Game loop
while True:
    display_board(state)


    if is_goal(state):
        print("\nGoal state reached!")
        break


    move = input("Move blank (up/down/left/right): ").lower()
    move_tile(state, move)
```

# Tic Tac Toe (Easy Version)

```python
board = [" " for _ in range(9)]

def print_board():
    print(board[0],"|",board[1],"|",board[2])
    print("--+---+--")
    print(board[3],"|",board[4],"|",board[5])
    print("--+---+--")
    print(board[6],"|",board[7],"|",board[8])

def check_win(player):
    win_cond = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]

    for cond in win_cond:
        if board[cond[0]] == board[cond[1]] == board[cond[2]] == player:
            return True
    return False

def play():
    player = "X"
    for i in range(9):
        print_board()
        pos = int(input("Enter Position(0-8)"))
        if board[pos] == " ":
            board[pos] = player
            if check_win(player):
                print_board()
                print(f"Player {player} Wins!")
                return
            player = "O" if player == "X" else "X"
        else:
            print("Invalid Move")

    print("Draw")

play()        
```

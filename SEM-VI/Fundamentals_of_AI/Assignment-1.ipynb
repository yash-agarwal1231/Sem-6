# 1.Breadth First Search(BFS)

```python
from collections import deque
graph = {
    'A' : ['B','C'],
    'B' : ['D'],
    'C' : ['F'],
    'D' : ['G'],
    'F' : [],
    'G' : []
}

def bfs(start):
    visited = []
    queue = deque([start])
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])
    return visited
```

```python
print("\nBreadth First Search Traversal:- \n")
print(bfs('A'))
```

# 2.Depth First Search (DFS)

```python
graph = {
    'Root' : ['Folder1','Folder2'],
    'Folder1' : [],
    'Folder2' : []
}
visited = []
def dfs(node):
    visited.append(node)
    for child in graph[node]:
        dfs(child)
    return visited
```

```python
print("\nDepth First Search Traversal is :-\n")
print(dfs('Root'))
```

# 1. A* Algorithm

```python
from heapq import heappush,heappop

graph = {
    'S' : [('A',1),('B',4)],
    'A' : [('D',5)],
    'B' : [('G',1)],
    'D' : [('G',3)],
    'G' : []
}

heuristic = {
    'S' : 7,
    'A' : 6,
    'B' : 5,
    'D' : 2,
    'G' : 0
}

def aStar(start,goal):
    pq=[]
    heappush(pq,(heuristic[start],0,start,[start]))
    while pq:
        f,cost,node,path = heappop(pq)
        if node == goal:
            return path,cost
        
        for neighbour, edge_cost in graph[node]:
            g = cost + edge_cost
            h = heuristic[neighbour]
            heappush(pq,(g+h,g,neighbour,path+[neighbour]))
```

```python
print("\nA* Algorithm\n")
print("Shortest Path from S to G is: \n")
print(aStar('S','G'))
```

# 2. Best First Search

```python
graph = {
    'Start' : ['X','Y'],
    'Y' : ['Z'],
    'Z' : ['T'],
    'T' : []
}

heuristic = {
    'Start' : 2,
    'X' : 6,
    'Y' : 4,
    'Z' : 2,
    'T' : 0
}

def best_first_search(start,goal):
    pq = []
    visited = []
    heappush(pq,(heuristic[start],start))
    while pq:
        h,node = heappop(pq)
        visited.append(node)

        if node == goal:
            return visited
        
        for neighbour in graph[node]:
            heappush(pq,(heuristic[neighbour],neighbour))
```

```python
print(best_first_search('Start','T'))
```

# 3. Water-Jug Problem

```python
from collections import deque

def water_jug():
    capA=4
    capB=3
    target=(2,0)
    q=deque()
    q.append((0,0))
    visited=set()
    visited.add((0,0))
    parent={}
    parent[(0,0)]=None    
    while q:
        a,b=q.popleft()
        if(a,b)==target:
            path=[]
            curr=(a,b)
            while curr:
                path.append(curr)
                curr=parent[curr]
       
            path.reverse()
            print("Water Jug Problem - Path:")
            step = 1
            for state in path:
                print(f"Step {step} : {state}")
                step += 1
            return 
        next_states=[
        (capA,b),
        (a,capB),
        (0,b),
        (a,0),
        (a-min(a,capB-b),b+min(a,capB-b)),
        (a+min(b,capA-a),b-min(b,capA-a))]
        for state in next_states:
            if state not in visited:
                visited.add(state)
                parent[state]=(a,b)
                q.append(state)
water_jug()
```

# 1. 8-Queens Problem

```python
def safe(board, row, col):
    for i in range(row):
        if board[i] == col or board[i] - i == col - row or board[i] + i == col + row:
            return False
    return True


def solve_queens(row, board, n):
    if row == n:
        print_solution(board)
        return True 

    for col in range(n):
        if safe(board, row, col):
            board[row] = col
            if solve_queens(row + 1, board, n):
                return True
            board[row] = -1


    return False


def print_solution(board):
    n = len(board)
    for i in range(n):
        for j in range(n):
            print("Q" if board[i] == j else ".", end=" ")
        print()


n = 8
board = [-1] * n
solve_queens(0, board, n)
```

# 2. Hill Climb Algorithm

```python
plateaus = {
    'BaseCamp(A)': 3,
    'PineHill(B)': 5,
    'EaglePeak(C)': 8,
    'RockyRidge(D)': 4
}


trail = ['BaseCamp(A)', 'PineHill(B)', 'EaglePeak(C)', 'RockyRidge(D)']
current = 'BaseCamp(A)'


for next_plateau in trail:
    #print(next_plateau)
    if plateaus[next_plateau] > plateaus[current]:
        print("Visited : ",current)
        current = next_plateau
        


print("Highest plateau reached:", current)
print("Height:", plateaus[current])
```

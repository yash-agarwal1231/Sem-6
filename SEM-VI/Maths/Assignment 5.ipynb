# Q2

```python
x,y = symbols('x,y')
T = Matrix([[1,3,0],[-2,2,0],[0,0,1]])
T_inv = T.inv()

L = Line(x-y+4)
p = L.points
a = p[0]
b = p[1]

a1 = a.transform(T_inv)
b1 = b.transform(T_inv)

Res = Line(a1,b1)
Res.equation()
```

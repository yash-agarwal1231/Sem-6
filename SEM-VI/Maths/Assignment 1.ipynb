```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5,100)
y = x**2
plt.plot(x,y,label = "x^2", color = 'r')
plt.title("Graph of x^2")
plt.xlabel("X-axis")
plt.ylabel("Y-label")
plt.legend()
plt.show()
```

## Q1 - 3D Plot

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
z = X**2 + Y**2
plt.contourf(X,Y,z)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## Q2

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
z = np.sin(X) + np.cos(Y)
plt.contourf(X,Y,z)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## Q4 : Sales - Product A  v\s  Product B

```python
months = ["Jan","Feb","Mar","Apr","May","Jun"]
product_A = [50,60,55,70,65,80]
product_B = [30,40,45,50,60,75]

plt.plot(months, product_A, color = "g", label = "Product A")
plt.plot(months, product_B, color = "c", label = "Product B")
plt.xlabel("Months")
plt.ylabel("Product Sales")
plt.title("Product Sales")
plt.legend()
plt.show()
```

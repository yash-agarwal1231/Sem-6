# Q1 - Co-Linear

### Q1


```python
from sympy import *
import numpy as np
import math
```

```python
X = Point(2,-1,3)
Y = Point(2,-1,4)
Z = Point(2,-1,-7)

res = Point.is_collinear(X,Y,Z)
print("Are the Points Co-Linear ? ", res)
```

### Q2

```python
X = Point(0,-2,-7)
Y = Point(0,-2,-7)
Z = Point(0,-2,-7)

res = Point.is_collinear(X,Y,Z)
print("Are the Points Co-Linear ? ", res)
```

# Q2 - Co-Planar

### Q1

```python
X = Point(2,-1,3)
Y = Point(2,-1,4)
Z = Point(2,-1,-7)

res = Point.are_coplanar(X,Y,Z)
print("Are the Points Co-Planar ? ", res)
```

### Q2

```python
X = Point(0,-2,-7)
Y = Point(0,-2,-7)
Z = Point(0,-2,-7)

res = Point.are_coplanar(X,Y,Z)
print("Are the Points Co-Planar ? ", res)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi,50)
y = np.linspace(0,2*np.pi,50)
X, Y = np.meshgrid(x,y)
Z = np.exp(X**2 + Y**2)
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_wireframe(X, Y, Z, color = 'blue', linewidth = 0.9, rstride = 2, cstride = 2)
ax.contour3D(X, Y, Z, 50, cmap ='viridis')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
plt.show()
```

# Surface Plots

# Q1

```python
x = np.linspace(0,2,100)
y = np.linspace(0,2,100)
X, Y = np.meshgrid(x,y)
Z = X * np.exp(Y)
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z, color = 'gold', linewidth = 0.7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
ax.set_title('3D Surface Plot of f(x,y) = x.e^y')
plt.show()
```

# SET A

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# Q1

```python
transactions = [['Bread','Milk'],
                ['Bread','Diaper','Beer','Eggs'],
                ['Milk','Diaper','Beer','Coke'],
                ['Bread','Milk','Diaper','Beer'],
                ['Bread','Milk','Diaper','Coke']]
transactions
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns = te.columns_)

freq_items = apriori(df, min_support = 0.4, use_colnames = True)
rules = association_rules(freq_items, metric = 'lift', min_threshold = 1.26)

print(freq_items)
print(rules[['antecedents','consequents','support','confidence','lift','conviction','leverage']])

```

# Q2

```python
transactions2 = [['Rice','Oil','Salt'],
                 ['Rice','Oil'],
                 ['Oil','Salt'],
                 ['Rice','Salt'],
                 ['Rice','Oil','Salt']]

te = TransactionEncoder()
te_array2 = te.fit(transactions2).transform(transactions2)
df = pd.DataFrame(te_array2, columns = te.columns_)

freq_items1 = apriori(df, min_support = 0.4, use_colnames = True)
rules = association_rules(freq_items1, metric = 'confidence', min_threshold = 0.6)
print(freq_items1)
print(rules[['antecedents','consequents','support','confidence','lift']])
```
